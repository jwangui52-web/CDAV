```python
for i in range(1,11):
  for j in range(1,11):
    print(f"{i}*{j}={i*j}")
  print()

```

```python
print("\t",end="")
#print column heading
for i in range(1,11):
 print(i, end="\t")
print()

#print rows
for i in range(1,11):
 print(i,  end="\t")
 #row heading
for j in range(1,11):
  print(i * j, end="\t")
  print()
```
